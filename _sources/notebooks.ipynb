```python
dataset2 ="https://raw.githubusercontent.com/aceburakhol/dataset/main/Bank%20Customer%20Churn%20Prediction.csv"
data2=pd.read_csv(dataset2)
```

```python
data2.head(100)
```

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
products_number = data2.products_number
credit_card = data2.credit_card
active_member = data2.active_member
customer_id = data2.customer_id
credit_score = data2.credit_score
```

```python
pn = products_number.values.reshape(-1,1)
cc = credit_card.values.reshape(-1,1)
am = active_member.values.reshape(-1,1)
ci = customer_id.values.reshape(-1,1)
cs = credit_score.values.reshape(-1,1)
```

```python
est = KBinsDiscretizer(n_bins=4, encode='ordinal', strategy='uniform')
```

```python
est.fit(pn)
est.fit(cc)
est.fit(am)
```

```python
aa = est.transform(pn)
ab = est.transform(cc)
ac = est.transform(am)
```

```python
a1 = pd.DataFrame(aa,columns=['products_number'])
a2 = pd.DataFrame(ab,columns=['credit_card'])
a3 = pd.DataFrame(ac,columns=['active_member'])
a4 = pd.DataFrame(ci,columns=['customer_id'])
a5 = pd.DataFrame(cs,columns=['credit_score'])
```

```python
datalist = [a1, a2, a3, a4, a5]
pd.concat(datalist, axis=1)
```
